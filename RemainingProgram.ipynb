```python
#1 Write a function that takes as parameter a list of mixed data and returns segregated list of numeric and strings

def segregate_data(mixed_list):
    numbers = []
    strings = []
    
    for item in mixed_list:
        if isinstance(item, (int, float)):
            numbers.append(item)
        elif isinstance(item, str):
            strings.append(item)
    
    return numbers, strings

data = [12, "apple", 3.14, "banana", 42, "cherry", -5]
nums, strs = segregate_data(data)

print("Numbers:", nums)
print("Strings:", strs)
```

```python
#2 Write a function that takes a list of numbers as parameter and returns a list of all zeros repositioned at the beginning

def zeros(num_list):
    zeros = [0] * num_list.count(0)   # list of all zeros
    non_zeros = [x for x in num_list if x != 0]  # rest of the numbers
    return zeros + non_zeros

numbers = [1, 0, 2, 0, 3, 4, 0, 5]
result = zeros(numbers)

print(result)
```

```python
#3 Write a function that takes a filename as a parameter and if its a valid fasta file returns the sequence and the header as strings

def fasta (filename):
    with open(filename, "r") as file:
        lines = file.readlines()

    if not lines or not lines[0].startswith(">"):
        return None, None   # not a valid FASTA file

    header = lines[0].strip()
    sequence = "".join(line.strip() for line in lines[1:] if line.strip())

    return header, sequence


# Example usage
header, seq = fasta("sequence.fasta")
print("Header:", header)
print("Sequence:", seq)
```

```python
#1 . Create a class Rectangle 
#Data members :
#length
#breadth
#Methods :
#getArea
#getPerimeter
#Show

class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def getArea(self):
        """Return the area of the rectangle"""
        return self.length * self.breadth

    def getPerimeter(self):
        """Return the perimeter of the rectangle"""
        return 2 * (self.length + self.breadth)

    def show(self):
        """Display rectangle details"""
        print(f"Rectangle: Length = {self.length}, Breadth = {self.breadth}")
        print(f"Area = {self.getArea()}")
        print(f"Perimeter = {self.getPerimeter()}")


# Example usage
rect = Rectangle(5, 3)
rect.show()
```

```python
#3 Create a class Sequence
#Data member :
#Seq
#Methods :

#Create a class DNASequence inheriting from the Sequence Class
#The constructor should validate the sequence before assigning
#Methods :
#GetComplement
#GetGC

class Sequence:
    def __init__(self, seq=""):
        self.Seq = seq


class DNASequence(Sequence):
    def __init__(self, seq):
        # Validate sequence contains only A, T, G, C (case-insensitive)
        seq = seq.upper()
        if all(base in "ATGC" for base in seq):
            super().__init__(seq)
        else:
            raise ValueError("Invalid DNA sequence: must contain only A, T, G, C")

    def GetComplement(self):
        """Return the complementary DNA sequence"""
        complement = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
        return ''.join(complement[base] for base in self.Seq)

    def GetGC(self):
        """Return the GC content percentage of the sequence"""
        gc_count = sum(1 for base in self.Seq if base in 'GC')
        return (gc_count / len(self.Seq)) * 100 if self.Seq else 0


# Example usage
dna = DNASequence("ATGCGT")
print("Original Sequence:", dna.Seq)
print("Complement:", dna.GetComplement())
print("GC Content (%):", dna.GetGC())
```

```python
#5 Create a class Distance
#Data members :
#Feet
#Inches
#Methods
#Show
#__add__
#__str__
#__eq__
#__ne__

class Distance:
    def __init__(self, feet=0, inches=0):
        # Normalize inches to be less than 12
        self.Feet = feet + inches // 12
        self.Inches = inches % 12

    def Show(self):
        """Display the distance"""
        print(f"{self.Feet} feet {self.Inches} inches")

    def __add__(self, other):
        """Add two Distance objects"""
        total_inches = (self.Feet + other.Feet) * 12 + self.Inches + other.Inches
        return Distance(0, total_inches)

    def __str__(self):
        """String representation of the Distance object"""
        return f"{self.Feet} feet {self.Inches} inches"

    def __eq__(self, other):
        """Check if two distances are equal"""
        return self.Feet == other.Feet and self.Inches == other.Inches

    def __ne__(self, other):
        """Check if two distances are not equal"""
        return not self.__eq__(other)


# Example usage
d1 = Distance(5, 8)
d2 = Distance(3, 9)

print("Distance 1:", d1)
print("Distance 2:", d2)

d3 = d1 + d2
print("Sum of distances:", d3)

print("Are d1 and d2 equal?", d1 == d2)
print("Are d1 and d2 not equal?", d1 != d2)
```

```python
#1 Create a module with the following functions
#	a. Write a function that takes a DNA sequence as a parameter and return the 
#          GC percent
#	b. Write a function that takes a DNA sequence as a parameter and returns the 
#           complement of the sequence
#    c. Write a function that takes a DNA sequence as a parameter and returns the 
#           RNA sequence
#Use the module in the main script

def get_gc_percent(dna_seq):
    gc_count = dna_seq.upper().count('G') + dna_seq.upper().count('C')
    total_bases = len(dna_seq)
    gc_percent = (gc_count / total_bases) * 100
    return gc_percent

def get_complement(dna_seq):
    complement_map = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
    complement_seq = ''
    for base in dna_seq.upper():
        complement_seq += complement_map[base]
    return complement_seq

def get_rna(dna_seq):
    rna_seq = dna_seq.upper().replace('T', 'U')
    return rna_seq
```
